```python
import numpy as np 
import pandas as pd 
from scipy.interpolate import CubicSpline

import plotly.express as px
import plotly.graph_objects as go
```

# 碳梯度和碳分布结果

```python
# csv数据读取为dataframe

csv_file = "./data/data.csv"
csv_data = pd.read_csv(csv_file, low_memory = False) #防止弹出警告
csv_df = pd.DataFrame(csv_data)
csv_df.head(5)
```

```python
fig = px.line(
    csv_df, 
    x="Distance from the surface(mm)", 
    y="Carbon content(wt%)", 
    color="Stage",
    width=809, height=500, 
    template = "plotly_white", # ggplot2 plotly_white seaborn none simple_white
    )

fig.update_xaxes(gridcolor='black', griddash='dash', minor_griddash="dot")

fig.update_layout(
    font_family="Times New Roman",
    font_color="black",
    title_font_family="Times New Roman",
    title_font_color="red", 
    legend_title_font_color="black",
    legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=0.8),
)

fig.add_hline(y=0.2)
fig.add_vrect(x0=0.8, x1=1.2,
              fillcolor="red", opacity=0.15, line_width=0,
              annotation_text="Effective Depth", annotation_position="top",)

fig.show()
```
